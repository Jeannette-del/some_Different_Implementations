#*Exercise 1:* 
Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties: â€“

* X label name = Month Number
* Y label name = Total profit
* The line plot graph should look like this.

```
plt.figure(figsize=(10,6))
X = df['month_number']
Y = df['total_profit']
plt.plot(X,Y,linewidth=5)
plt.xlabel('Month Number',size=20)
plt.ylabel('Total profit',size=20)
plt.title('Company profit per Month',size=20)
plt.show()
```

#*Exercise Question 5: Read face cream and facewash product sales data and show it using the bar chart*

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

![](https://pynative.com/wp-content/uploads/2019/01/matplotlib_panadas_exercise_5_show_bar_chart.png)

```
plt.figure(figsize=(10,6))
u = df['facecream']
v = df['facewash']
barwidth = 0.4
x = np.arange(len(df))

plt.bar(x-barwidth/2, u, barwidth, label='Face Cream Sales Data')
plt.bar(x+barwidth/2, v, barwidth,label='Facewash Sales Data')
plt.xlabel('Month Number',size=20)
plt.ylabel('Sales units in Number',size=20)
plt.title('Facewash and Facecream sales data',size=20)
plt.legend(loc='upper left')
plt.grid(ls='--')
plt.show()
```

#*Exercise Question 8: Read all product sales data and show it using the stack plot*

The Stack plot should look like this.

![](https://pynative.com/wp-content/uploads/2019/01/matplotbib_pandas_exercise10_show_stack_plot.png)

```
plt.figure(figsize=(10,6))
y1 = df[df.columns[1:7][0]]
y2 =df[ df.columns[1:7][1]]
y3 =df[ df.columns[1:7][2]]
y4 = df[df.columns[1:7][3]]
y5 = df[df.columns[1:7][4]]
y6 = df[df.columns[1:7][5]]

colors = ['purple','cyan','red','black','green','olive']
labels = df.columns[1:7]

# y = df[y_cols]
y_stack = np.vstack((y1, y2, y3, y4, y5, y6))
plt.stackplot(df['month_number'], y_stack, colors=colors, labels=labels)
plt.xlabel('MOnth NUmber',size=20)
plt.ylabel('Sales Units in Number',size=20)
plt.legend(loc='upper left')
plt.title('All products sales data using stack plot',size=20)
plt.show()
# z = np.vstack()
```
